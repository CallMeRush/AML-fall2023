### Drop trailing NaN

```python
def drop_trailing_na(df: pd.DataFrame):
    return [df.loc[i].dropna().to_numpy() for i in range(df.shape[0])]

X_train = drop_trailing_na(X_train_df)
y_train = y_train_df['y'].to_numpy()
X_test = drop_trailing_na(X_test_df)
```
